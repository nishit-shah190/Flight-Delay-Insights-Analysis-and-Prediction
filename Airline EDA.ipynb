```python
# Prediction
features = [
    "CRS_DEP_TIME", "DEP_DELAY", "TAXI_OUT", "DISTANCE", 
    "DELAY_DUE_CARRIER", "DELAY_DUE_WEATHER", "DELAY_DUE_NAS", 
    "DELAY_DUE_SECURITY", "DELAY_DUE_LATE_AIRCRAFT"
]
target = "ARR_DELAY"
df_late_arrivals= df[df["ARR_DELAY"]>0].copy()
df_late_arrivals.loc[:, features] = df_late_arrivals[features].fillna(0)
x_train, x_test, y_train, y_test = train_test_split(df_late_arrivals[features], df_late_arrivals[target], test_size=0.2, random_state=42)
```

```python
df_late_arrivals.shape[0]
```

```python
model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model.predict(x_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
r2 = r2_score(y_test,y_pred)

print(f"Model Performance")
print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2: .2f} ")
```

```python
importance = pd.Series(model.coef_, index=features).sort_values(ascending=False)
print("Feature Importance:\n", importance)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x_train, y_train, cv=5, scoring='r2')
print("Cross-Validation R2 score:", scores)
print("Mean R2", scores.mean())
```

```python
from sklearn.metrics import mean_absolute_error

# Get predictions on test set
y_pred = model.predict(x_test)

# Compute MAE
mae = mean_absolute_error(y_test, y_pred)

# Compute Mean of Actual Values
mean_actual = y_test.mean()

# Compute Accuracy
accuracy = 1 - (mae / mean_actual)
print("Regression Accuracy:", accuracy)
```

```python
features = [
    "CRS_DEP_TIME", "DEP_DELAY", "TAXI_OUT", "DISTANCE", 
    "DELAY_DUE_CARRIER", "DELAY_DUE_WEATHER", "DELAY_DUE_NAS", 
    "DELAY_DUE_SECURITY", "DELAY_DUE_LATE_AIRCRAFT"
]
target_cancelled = "CANCELLED"  # Binary target (0 = Not Canceled, 1 = Canceled)

```

```python
df_classification = df.copy()
df_classification[features] = df_classification[features].fillna(0)
```

```python
df_classification
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_classification[features], df_classification[target_cancelled], test_size=0.2, random_state=42, stratify = df_classification[target_cancelled])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
model = LogisticRegression(class_weight="balanced", random_state=42)
model.fit(x_train, y_train)

rf_model= RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python


feature_importance = pd.Series(rf_model.feature_importances_, index=features).sort_values(ascending=False)
print("Feature Importance:\n", feature_importance)
```
